```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.arange(-5.0,5.0,0.1)
print("size of x Data is",x.size)
x
```

```python
plt.plot(x)
```

```python
y=2*(x)+3
y_noise=0.2*np.random.normal(size=x.size)
ydata=y+y_noise

plt.figure(figsize=(10,6))
plt.scatter(x,ydata)
plt.plot(x,y,'r')
plt.title("Linear Regression Line")
plt.xlabel('independent variable')
plt.ylabel('Dependent variable')
plt.show()
```

```python
x=np.arange(-5.0,5.0,0.1)
print("size of x Data is",x.size)
x
y=1*(x**3)+1*(x**2)+1*x+3
y_noise=0.2*np.random.normal(size=x.size)
ydata=y+y_noise

plt.figure(figsize=(10,6))
plt.scatter(x,ydata)
plt.plot(x,y,'r')
plt.title("Polinomial Equation ")
plt.xlabel('Independent variable')
plt.ylabel('Dependent variable')
plt.show()
```

```python
x=np.arange(-5.0,5.0,0.1)
print("size of x Data is",x.size)
x
y=np.log(1*(x**3)+1*(x**2)+1*x+3)
y_noise=0.2*np.random.normal(size=x.size)
ydata=y+y_noise

plt.figure(figsize=(10,6))
plt.scatter(x,ydata)
plt.plot(x,y,'r')
plt.title("Polinomial Equation ")
plt.xlabel('Independent variable')
plt.ylabel('Dependent variable')
plt.show()
```

```python
x=np.arange(-5.0,5.0,0.1)
print("size of x Data is",x.size)
x
y=np.power(x,2)
y_noise=0.2*np.random.normal(size=x.size)
ydata=y+y_noise

plt.figure(figsize=(10,6))
plt.scatter(x,ydata)
plt.plot(x,y,'r')
plt.title("quadratic or plob")
plt.xlabel('Independent variable')
plt.ylabel('Dependent variable')
plt.show()
```

```python
x=np.linspace(-9,9,400)
y=np.linspace(-5,5,400)
x,y=np.meshgrid(x,y)
```

```python
def axes():
    plt.figure(figsize=(16,8))
    plt.axhline(0,alpha=0.9)
    plt.axvline(0,alpha=0.9)
```

```python
a=.2
b=.1
axes()
plt.contour(x,y,(x**2/a**2-y**2/b**2),[1],colors='y')
plt.show()
```

```python
axes()
plt.contour(x,y,(y**2-4*a*x),[0],colors='y')
plt.show()
```

```python
x=np.arange(-5.0,5.0,0.1)
print("size of x Data is",x.size)
x
y=np.exp(x)
y_noise=5*np.random.normal(size=x.size)
ydata=y+y_noise

plt.figure(figsize=(10,6))
plt.scatter(x,ydata)
plt.plot(x,y,'r')
plt.title("Exponential")
plt.xlabel('Independent variable')
plt.ylabel('Dependent variable')
plt.show()
```

```python
x=np.arange(-5.0,5.0,0.1)
print("size of x Data is",x.size)
x
y=1-4/(1+np.power(3,x-2))
y_noise=0.5*np.random.normal(size=x.size)
ydata=y+y_noise

plt.figure(figsize=(10,6))
plt.scatter(x,ydata)
plt.plot(x,y,'r')
plt.title("sigmoid")
plt.xlabel('Independent variable')
plt.ylabel('Dependent variable')
plt.show()
```
